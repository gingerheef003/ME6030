# Orthogonal Polynomial

The basis of the vector space is $1,x,x^2,x^3,\dots$

Inner Product: $<f,g> \int_{-1}^{1}f(x)g(x)dx$

We get the Legendre Polynomials


```python
def inner_product(f,g):
    '''Inner product between two functions'''
    return integrate(f*g, (x,-1,1))

def norm(f):
    '''Norm of the function'''
    return sqrt(inner_product(f,f))
```

# Gramm\-Schmidt Orthonormallization



```python
L1p = 1
L1 = L1p / norm(L1p)

L2p = x - inner_product(x,L1)*L1
L2 = L2p / norm(L2p)

L3p = x^2 - inner_product(x^2,L1)*L1 - inner_product(x^2,L2)*L2
L3 = L3p / norm(L3p)

L4p = x^3 - inner_product(x^3,L1)*L1 - inner_product(x^3,L2)*L2 - inner_product(x^3,L3)*L3
L4 = L4p / norm(L4p)

show(L4)

p1 = plot(L1, (-1,1), color='blue')
p2 = plot(L2, (-1,1), color='red')
p3 = plot(L3, (-1,1), color='green')
p4 = plot(L4, (-1,1), color='black')
show(p1+p2+p3+p4)
```

The basis of the vector space is $1,x,x^2,x^3,\dots$

Inner Product: $\displaystyle <f,g> \int_{-1}^{1}f(x)g(x)\frac{1}{\sqrt{1 - x^2}}dx$

We get the Chebysev Polynomials


```python
def inner_product(f,g):
    '''Inner product between two functions'''
    return integrate(f*g*1/sqrt(1 - x^2), (x,-1,1))

def norm(f):
    '''Norm of the function'''
    return sqrt(inner_product(f,f))
```

```python
L1p = 1
L1 = L1p / norm(L1p)

L2p = x - inner_product(x,L1)*L1
L2 = L2p / norm(L2p)

L3p = x^2 - inner_product(x^2,L1)*L1 - inner_product(x^2,L2)*L2
L3 = L3p / norm(L3p)

L4p = x^3 - inner_product(x^3,L1)*L1 - inner_product(x^3,L2)*L2 - inner_product(x^3,L3)*L3
L4 = L4p / norm(L4p)

show(L4)

p1 = plot(L1, (-1,1), color='blue')
p2 = plot(L2, (-1,1), color='red')
p3 = plot(L3, (-1,1), color='green')
p4 = plot(L4, (-1,1), color='black')
show(p1+p2+p3+p4)
```

```python
def get_polu(ip, num):
    L = []
    for i in range(num):
        Lp = x^i
        for j in range(i):
            Lp -= ip(x^i,x^j)*x^j
        Lp = Lp / norm(Lp)
        L.append(Lp)
    return L
```
